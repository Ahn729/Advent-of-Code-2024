```python
with open("input.txt") as f:
    lines = [line.strip() for line in f.readlines()]
```

```python
try:
    from tqdm import tqdm
except ImportError:
    tqdm = lambda x: x
```

```python
stones = [int(stone) for stone in lines[0].split()]
```

```python
def apply_blink_to_stone(stone: int) -> list[int]:
    if stone == 0:
        return [1]
    else:
        s_stone = str(stone)
    if not len(s_stone) % 2:
        return [int(s_stone[:len(s_stone)//2]), int(s_stone[len(s_stone)//2:])]
    return [stone * 2024]

def apply_blink_to_stones(stones: list[str]) -> list[str]:
    new_stones = []
    for stone in stones:
        new_stones.extend(apply_blink_to_stone(stone))
    return new_stones
```

```python
for _ in tqdm(range(25)):
    stones = apply_blink_to_stones(stones)
```
